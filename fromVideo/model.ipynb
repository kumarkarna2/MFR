```python
import os
import cv2
import face_recognition
from datetime import datetime
```

```python
known_criminals_folder = "Test/Assault/"
database_folder = "database"
known_encodings = []
known_names = []
```

```python
def load_known_criminals():
    for file_name in os.listdir(known_criminals_folder):
        image_path = os.path.join(known_criminals_folder, file_name)
        name = os.path.splitext(file_name)[0]
        image = face_recognition.load_image_file(image_path)

        # Handle the case where a face cannot be encoded
        try:
            encoding = face_recognition.face_encodings(image)[0]
            known_encodings.append(encoding)
            known_names.append(name)
        except IndexError:
            print(f"Error: Could not encode face in {file_name}. Skipping.")
```

```python
def store_in_database(name):
    timestamp = datetime.now().strftime("%Y%m%d%H%M%S")
    file_name = f"{name}_{timestamp}.jpg"
    save_path = os.path.join(database_folder, file_name)
    cv2.imwrite(save_path, frame)
    print(f"{name} stored in the database successfully.")
```

```python
def match_and_store_in_database(video_path):
    cap = cv2.VideoCapture(video_path)
    while True:
        ret, frame = cap.read()
        if not ret:
            break

        face_locations = face_recognition.face_locations(frame)
        face_encodings = face_recognition.face_encodings(frame, face_locations)

        for face_encoding in face_encodings:
            matches = face_recognition.compare_faces(known_encodings, face_encoding)
            name = "Unknown"

            if True in matches:
                matched_indices = [
                    index for index, match in enumerate(matches) if match
                ]
                first_match_index = matched_indices[0]
                name = known_names[first_match_index]
                store_in_database(name)

            for top, right, bottom, left in face_locations:
                cv2.rectangle(frame, (left, top), (right, bottom), (237, 255, 32), 2)
                cv2.putText(
                    frame,
                    f"Matched Person: {name}",
                    (left - 40, top - 40),
                    cv2.FONT_HERSHEY_SIMPLEX,
                    0.9,
                    (158, 49, 255),
                    2,
                )

        cv2.imshow("Match and Store in Database", frame)
        if cv2.waitKey(1) & 0xFF == ord("q"):
            break

    cap.release()
    cv2.destroyAllWindows()
```
